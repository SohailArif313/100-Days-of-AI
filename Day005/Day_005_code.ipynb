### `Q1:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
# write your code here

# List of tuples with repeated first elements
test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]

# Dictionary to group values by their key
l1 = {}

for num1, num2 in test_list:
    
    # If key doesn't exist, create a new list with the value
    if num1 not in l1:
        l1[num1] = [num2]
    else:
        # If key exists, append the value to its list
        l1[num1].append(num2)

# Merge each key with its grouped values into a single tuple
ls = []
for key, value in l1.items():
    ls.append((key, *value))  # * unpacks the value list into the tuple

print(ls)
````

### `Q2:` Multiply Adjacent elements (both side) and take sum of right and lest side multiplication result.


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication :

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
# write your code here

l1 = (1, 5, 7, 8, 10)
result = []

for i in range(len(l1)):
    
    # First element: multiply with next element only
    if i == 0:
        result.append(l1[i] * l1[i+1])
    
    # Last element: multiply with previous element only
    elif i == len(l1) - 1:
        result.append(l1[i] * l1[i-1])
    
    # Middle elements: multiply with both previous and next element
    else:
        result.append(l1[i] * l1[i-1] + l1[i] * l1[i+1])

# Convert result list back to tuple
print(tuple(result))
````

### `Q4`: Count no of tuples, list and set from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
# write your code here

# A list containing mixed data types (sets, tuples, lists)
list1 = [{'hi', 'bye'}, {'Geeks', 'forGeeks'}, ('a', 'b'), ['hi', 'bye'], ['a', 'b']]

# Counters for each data type
list_count = 0
set_count = 0
tup_count = 0

# Loop through each element and check its type
for i in list1:
    
    if type(i) == list:
        list_count += 1    # Increment list counter
        
    elif type(i) == set:
        set_count += 1     # Increment set counter
        
    else:
        tup_count += 1     # Increment tuple counter

# Print the final counts
print(f"List-{list_count}\nSet-{set_count}\nTuples-{tup_count}")
````

### `Q5`: Shortlist Students for a Job role
Ask user to input students record and store in tuples for each record. Then Ask user to input three things he wants in the candidate- Primary Skill, Higher Education, Year of Graduation.

Show every students record in form of tuples if matches all required criteria.

It is assumed that there will be only one primry skill.

If no such candidate found, print `No such candidate`

`Input:`
```
Enter No of records- 2
Enter Details of student-1
Enter Student name- Manohar
Enter Higher Education- B.Tech
Enter Primary Skill- Python
Enter Year of Graduation- 2022
Enter Details of student-2
Enter Student name- Ponian
Enter Higher Education- B.Sc.
Enter Primary Skill- C++
Enter Year of Graduation- 2020

Enter Job Role Requirement
Enter Skill- Python
Enter Higher Education- B.Tech
Enter Year of Graduation- 2022
```

`Output`
```
('Manohar', 'B.tech', 'Python', '2022')
```


````python
# write your code here
# Dataset to store all student records
company_dataset = []

while True:
    print("For exit enter 'break'")
    choice = input("Enter choice (break/continue): ")

    # Exit the loop if user enters 'break'
    if choice == "break":
        break

    students = int(input("Enter number of students: "))

    # Loop to collect details for each student
    for i in range(students):
        print(f"\nEnter Student {i+1} Details")
        name         = input("Enter Student name: ")
        education    = input("Enter Student Education: ")
        primary_skill = input("Enter Primary Skill: ")
        year_of_grad = int(input("Enter Year of Graduation: "))

        # Store student details as a tuple in the dataset
        company_dataset.append((name, education, primary_skill, year_of_grad))

    # Get job role requirements from user
    print("\nEnter Job Role Requirements")
    req_skill     = input("Enter Required Skill: ")
    req_education = input("Enter Required Education: ")
    req_year      = int(input("Enter Required Graduation Year: "))

    print("\nQualified Students:")

    # Flag to track if any matching student is found
    found = False

    # Loop through all students to find matching candidates
    for student in company_dataset:
        s_name, s_edu, s_skill, s_year = student

        # Check if student matches all job requirements
        if (s_skill == req_skill) and (s_edu == req_education) and (s_year == req_year):
            print(f"({s_name}, {s_edu}, {s_skill}, {s_year})")
            found = True

    # If no match found, display message
    if not found:
        print("No such candidate")
````

### `Q1:` Write a program to find set of common elements in three lists using sets.
```
Input : ar1 = [1, 5, 10, 20, 40, 80]
        ar2 = [6, 7, 20, 80, 100]
        ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

Output : [80, 20]
```

````python
# write your code here

# Three lists with some common elements
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

# Find common elements in all three lists using set intersection
result = set(ar1) & set(ar2) & set(ar3)

# Convert result back to list and print
print(list(result))
````

### `Q2:` Write a program to count unique number of vowels using sets in a given string. Lowercase and upercase vowels will be taken as different.

`Input:`
```
Str1 = "hands-on data science mentorship progrAm with live classes at affordable fee only on CampusX"
```
`Output:`
```
No of unique vowels-6
```

````python
# write your code her
# Given input string
Str1 = "hands-on data science mentorship progrAm with live classes at affordable fee only on CampusX"

# Define all vowels (lowercase and uppercase are different)
vowels = set("aeiouAEIOU")

# Find unique vowels present in the string using set intersection
unique_vowels = set(Str1) & vowels

# Print the count of unique vowels
print(f"No of unique vowels-{len(unique_vowels)}")
````

### `Q3:` Write a program to Check if a given string is binary string of or not.

A string is said to be binary if it's consists of only two unique characters.

Take string input from user.

```
Input: str = "01010101010"
Output: Yes

Input: str = "1222211"
Output: Yes

Input: str = "Campusx"
Output: No
```

````python
# write your code here
str1 = "01010101010"

# Find unique characters in the string using set
unique_chars = set(str1)

# If string has exactly 2 unique characters, it is a binary string
if len(unique_chars) == 2:
    print("Yes")
else:
    print("No")
````
