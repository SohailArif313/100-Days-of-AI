### `Q-1:` Find the nearest element in the array to a given integer.
```
Ex:-
a=23 and array - [10 17 24 31 38 45 52 59].
Nearest element is 24
```
Hint: Read about this function `argmin()`

````python
import numpy as np
````

````python
def nearest_element(arr, a):
    idx = np.abs(arr - a).argmin()  # index of smallest difference
    return arr[idx]


arr = np.array([10, 17, 24, 31, 38, 45, 52, 59])
a = 23

print(f"Array:   {arr}")
print(f"Given:   {a}")

print(f"Nearest: {nearest_element(arr, a)}")
````

### `Q-2:` Replace multiples of 3 or 5 as 0 in the given array.
```
arr=[1 2 3 4 5 6 7 9]

result-> [1 2 0 4 0 0 7 0]
```

````python
arr = np.array([1, 2, 3, 4, 5, 6, 7, 9])

arr[(arr % 3 == 0) | (arr % 5 == 0)] = 0  # replace multiples of 3 or 5 with 0

print(arr)
````

````python
# Part 1: Double elements at given indexes
arr = np.arange(10)
indexes = [0, 3, 4, 9]

arr[indexes] *= 2  # double elements at given indexes

print("Part 1:", arr)


# Part 2: Internal repetition using fancy indexing
array = np.array([1, 2, 3])
n = len(array)  # repetition = length of array = 3

fancy_index = np.repeat(np.arange(n), n)  # [0,0,0, 1,1,1, 2,2,2]
result = array[fancy_index]               # fetch elements using fancy index

print("Part 2:", result)
````

### `Q-4:` Your are given an array which is havig some nan value. You job is to fill those nan values with most common element in the array.
```
arr=np.array([[1,2,np.nan],[4,2,6],[np.nan,np.nan,5]])

```

````python
arr = np.array([[1, 2, np.nan],
                [4, 2,      6],
                [np.nan, np.nan, 5]])

flat        = arr[~np.isnan(arr)]               # remove nan values
values, counts = np.unique(flat, return_counts=True)  # count each element
most_common = values[counts.argmax()]           # element with highest count

arr[np.isnan(arr)] = most_common               # replace nan with most common

print("Most Common Element:", most_common)
print("\nFilled Array:")
print(arr)
````

````python
arr = np.array([[3,  2,      np.nan, 1     ],
                [10, 12,     10,     9     ],
                [5,  np.nan, 1,      np.nan]])

# Boolean matrix — True where nan
print("Missing Data (Boolean Matrix):")
print(np.isnan(arr))

# Fill nan with 0
print("\nAfter Filling NaN with 0:")
print(np.nan_to_num(arr))
````

### `Q-6:` Given two arrays, X and Y, construct the Cauchy matrix C.
`Cij =1/(xi - yj)`

http://en.wikipedia.org/wiki/Cauchy_matrix
```
x = numpy.array([1,2,3,4]).reshape((-1, 1)
y = numpy.array([5,6,7])
```

````python
x = np.array([1, 2, 3, 4]).reshape((-1, 1))  # shape (4,1)
y = np.array([5, 6, 7])                       # shape (3,)

C = 1 / (x - y)  # broadcasting: (4,1) - (3,) = (4,3)

print("X:\n", x.flatten())
print("Y:\n", y)
print("\nCauchy Matrix:")
print(C)
````

### `Q-7:` Plot this below equation.

$$\large y = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

**Note: This equation is called tanh activation function. In deep learning, many times this function is used. If you find some difference between the sigmoid function and this tanh function, note that to your notebook.**

````python
import matplotlib.pyplot as plt
# Create x values
x = np.linspace(-5, 5, 100)

# Sigmoid formula
sigmoid = 1 / (1 + np.exp(-x))

# Tanh formula
tanh = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

# Draw both graphs
plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, tanh, label="Tanh")

# Add title
plt.title("Sigmoid vs Tanh")

# Name the axes
plt.xlabel("x")
plt.ylabel("Output")

# Show grid and labels
plt.grid()
plt.legend()

# Show graph
plt.show()
````

````python
# Output Range: Tanh gives outputs from -1 to 1, while sigmoid gives outputs from 0 to 1.
# Zero-Centered: Tanh is zero-centered because its output is 0 at x = 0, while sigmoid gives 0.5 at x = 0.
# Stronger Gradient: Tanh has a larger gradient than sigmoid, so it usually learns faster in shallow neural networks.
````

### `Q-9:` Write a program implement Boradcasting Rule to check if two array can be added or not.
Given tuples of shapes.
```
shape of a- (3,2,2)
shape of b- (2,2)

check_broadcast(a, b) -> return Boolean (True if can broadcasted, False other wise.)
```

````python
def check_broadcast(shape_a, shape_b):

    # Reverse both shapes
    a = shape_a[::-1]
    b = shape_b[::-1]

    # Compare dimensions one by one
    for da, db in zip(a, b):  # (2,2,3) ,(1,2,2)

        # If dimensions are different and neither is 1
        if da != db and da != 1 and db != 1:
            return False

    # All dimensions are valid
    return True


a = (3, 2, 2)
b = (2, 2)

result = check_broadcast(a, b)

print(f"shape a : {a}")
print(f"shape b : {b}")
print(f"can broadcast : {result}")
````
