### `Q-1:` Create a random 3x4 matrix with value between 0-100. And perform below tasks

    i. Sort this matrix. np.sort()
    ii. Sort this matrix based on values in 2nd column.
    iii. Sort this matrix based on max value in each row.
    iv. Sort based on elements value.


See examples:
```
arr =
    [[92 90 74]
    [ 6 63 93]
    [15 93 96]
    [70 60 48]]
```
```
i. np.sort
[[74 90 92]
 [ 6 63 93]
 [15 93 96]
 [48 60 70]]

ii. based on 2nd column
[[70 60 48]
 [ 6 63 93]
 [92 90 74]
 [15 93 96]]

iii. based on row max- descending
[[15 93 96]
 [ 6 63 93]
 [92 90 74]
 [70 60 48]]

iv. based on elements value
[[ 6 15 48]
 [60 63 70]
 [74 90 92]
 [93 93 96]]

```

````python
import numpy as np

arr = np.array([[92, 90, 74],
                [ 6, 63, 93],
                [15, 93, 96],
                [70, 60, 48]])
````

````python
def sort_rowwise(a):
    # sort each row internally
    return np.sort(a, axis=1)

def sort_by_col(a, col=1):
    # sort rows based on given column
    return a[a[:, col].argsort()]

def sort_by_row_max(a, ascending=False):
    # sort rows based on row max
    idx = np.max(a, axis=1).argsort()
    return a[idx[::-1]] if not ascending else a[idx]

def sort_all_elements(a):
    # sort all elements and reshape back
    return np.sort(a.flatten()).reshape(a.shape)

print("Original:\n", arr)
print("\ni.  row wise sort:\n",         sort_rowwise(arr))
print("\nii. sort by 2nd column:\n",    sort_by_col(arr, col=1))
print("\niii. sort by row max (desc):\n", sort_by_row_max(arr))
print("\niv. all elements sorted:\n",   sort_all_elements(arr))
````

### `Q-2:` There is an array of marks of 5 students in 4 subjects. Further you are asked to perform below task.
    i. Add marks every student of an extra subject in the same array.
    ii. Add two new students marks in respective 5 subjects.(one subject added in above task)
    iii. Add extra column with sum of all subjects(5-subjects) marks
    iv. Sort the array(non-ascending order) on total marks column--one added in above task. Show top 2 rows.

Note: Change dimension of arrays during concatenation or appending if required.

Given Array-
```
marks = [[13, 10,  9, 33],
       [63, 46, 90, 42],
       [39, 76, 13, 29],
       [82,  9, 29, 78],
       [67, 61, 59, 36]]

extra_subject = [41, 87, 72, 36, 92]
#Two extra students record-
rec1 = [77, 83, 98, 95, 89]
rec2 = [92, 71, 52, 61, 53]
```

````python
def add_subject(marks, extra_subject):
    # add extra subject as new column
    return np.concatenate([marks, extra_subject.reshape(-1, 1)], axis=1)

def add_students(marks, *records):
    # add new student rows
    new_rows = [r.reshape(1, -1) for r in records]
    return np.concatenate([marks, *new_rows], axis=0)

def add_totals(marks):
    # add total marks as last column
    totals = marks.sum(axis=1, keepdims=True)
    return np.concatenate([marks, totals], axis=1)

def top_students(marks, n=2):
    # sort by total (last column) descending, return top n
    sorted_marks = marks[marks[:, -1].argsort()[::-1]]
    return sorted_marks[:n]


# ── data ──
marks         = np.array([[13, 10,  9, 33],
                           [63, 46, 90, 42],
                           [39, 76, 13, 29],
                           [82,  9, 29, 78],
                           [67, 61, 59, 36]])

extra_subject = np.array([41, 87, 72, 36, 92])
rec1          = np.array([77, 83, 98, 95, 89])
rec2          = np.array([92, 71, 52, 61, 53])

# ── call functions ──
print("Original:\n", marks)

marks = add_subject(marks, extra_subject)
print("\ni.  After adding extra subject:\n", marks)

marks = add_students(marks, rec1, rec2)
print("\nii. After adding 2 students:\n", marks)

marks = add_totals(marks)
print("\niii. After adding total column:\n", marks)

print("\niv. Top 2 students:\n", top_students(marks))
````

### `Q-3:` Find unique arrays from a 2D array column wise and row wise.
```
arr = np.array([[1,2,3,3,1,1],
                [0,9,1,2,8,8],
                [1,2,3,8,8,8],
                [1,2,3,3,1,1]])
```
Expected Result-
```
Row Wise
[[0 9 1 2 8 8]
 [1 2 3 3 1 1]
 [1 2 3 8 8 8]]

Col Wise
[[1 1 2 3 3]
 [0 8 9 1 2]
 [1 8 2 3 8]
 [1 1 2 3 3]]
```

````python
import numpy as np

arr = np.array([[1, 2, 3, 3, 1, 1],
                [0, 9, 1, 2, 8, 8],
                [1, 2, 3, 8, 8, 8],
                [1, 2, 3, 3, 1, 1]])

# avoid repetition — functions banao
def unique_rows(a):
    return np.unique(a, axis=0)

def unique_cols(a):
    return np.unique(a, axis=1)

choice = input("Enter your choice (row / col / both): ").strip().lower()

if choice == "row":
    print("Row Wise:\n", unique_rows(arr))

elif choice == "col":
    print("Col Wise:\n", unique_cols(arr))

elif choice == "both":
    print("Row Wise:\n", unique_rows(arr))
    print("\nCol Wise:\n", unique_cols(arr))

else:
    print("Invalid input! Enter: row / col / both")
````

### `Q-5:` Get row numbers of NumPy array having element larger than X.
```
arr = [[1,2,3,4,5],
      [10,-3,30,4,5],
      [3,2,5,-4,5],
      [9,7,3,6,5]]

X = 6
```

````python
# given array and threshold
arr = np.array([[1,  2,  3,  4, 5],
                [10, -3, 30,  4, 5],
                [3,  2,  5, -4, 5],
                [9,  7,  3,  6, 5]])

X = 6

# check each row — True if any element > X
# np.where returns indices where condition is True
row_nums = np.where(np.any(arr > X, axis=1))[0]

print("Row numbers having element > 6:", row_nums)
````

### `Q-6:` How to convert an array of arrays into a flat 1d array?
arr1 = np.arange(3)
arr2 = np.arange(3,7)
arr3 = np.arange(7,10)


````python
def flatten_arrays(*arrays):
    # combine all arrays into one array of arrays
    arr = np.array(arrays, dtype=object)
    print("Original:", arr)

    # join all into one flat 1D array
    flat = np.concatenate(arr)
    print("Flattened:", flat)
    return flat

# call the function
flatten_arrays(np.arange(3), np.arange(3, 7), np.arange(7, 10))
````

### `Q-7:` You are given a array. You have to find the minimum and maximum array element and remove that from the array.

```python
import numpy as np

np.random.seed(400)
arr = np.random.randint(100, 1000, 200).reshape((1, 200))
```

````python
def remove_min_max(arr):
    print("Original shape:", arr.shape)
    print("Min:", arr.min(), "| Max:", arr.max())

    # flatten (1,200) → (200,)
    flat = arr.flatten()

    # find and remove min & max elements
    flat = np.delete(flat, np.where((flat == flat.min()) | (flat == flat.max())))

    print("\nNew shape:", flat.shape)
    print("New Min:", flat.min(), "| New Max:", flat.max())
    print("\nArray:\n", flat)
    return flat

# call the function
np.random.seed(400)
arr = np.random.randint(100, 1000, 200).reshape((1, 200))
remove_min_max(arr)
````

### `Q-8:` You are given an arrays. You have to limit this array's elements between 100 to 200. $arr ∈ [100, 700]$. So replace those values accordingly with the minimum and maximum value. Then sort the array and perform the cumulative sum of that array.

````python
def clip_sort_cumsum(arr, low=100, high=200):
    print("Original:\n", arr)

    # replace values below low with low, above high with high
    clipped = np.clip(arr, low, high)
    print(f"\nAfter clip ({low}–{high}):\n", clipped)

    # sort in ascending order
    sorted_arr = np.sort(clipped)
    print("\nSorted:\n", sorted_arr)

    # each element = sum of all previous elements + itself
    cumsum = np.cumsum(sorted_arr)
    print("\nCumulative sum:\n", cumsum)
    return cumsum

# call the function
np.random.seed(42)
arr = np.random.randint(100, 700, 20)
clip_sort_cumsum(arr)
````

### `Q-9:` You are given a array ($arr ∈ [0, 1]$). First you have round off the elements upto 3 decimal places and compare that
- 0th percentile == minimum value of the array
- 100th percentile == maximum value of the array
- also find the difference betwen 51th percenile and 50th percentile values

````python
def percentile_analysis(arr):
    # round to 3 decimal places
    arr = np.round(arr, 3)
    print("Array:\n", arr)

    # min and max
    print("\nMin value      :", arr.min())
    print("Max value      :", arr.max())

    # 0th and 100th percentile = min and max
    p0   = np.percentile(arr, 0)
    p100 = np.percentile(arr, 100)

    # 50th and 51st percentile
    p50  = np.percentile(arr, 50)
    p51  = np.percentile(arr, 51)

    print("\n0th  percentile:", p0,   "== Min?", p0   == arr.min())
    print("100th percentile:", p100, "== Max?", p100 == arr.max())

    print("\n50th percentile :", p50)
    print("51st percentile :", p51)

    # difference between 51st and 50th
    print("Difference      :", round(p51 - p50, 5))

# call the function
np.random.seed(42)
arr = np.random.random(20)
percentile_analysis(arr)
````
