### `Q-1`: You are given a function definition. There might be several issues on execution of this function. You are asked to do exception handling for diffrent errors that this function goes in to `without altering this function`. And print error text.



Function parameters `l -> list, s -> could be anything`

```
def function(l: list, s, **args):
    last_element = l[-1]
    
    l[int(s)]=10
    any_element = l[int(s)+10]
    l[s]=10
    
    res = sum(l)
    
    p = args['p']
    # print(p)
    return res/last_element * p + any_element

```
Check for different function calls:-

```
function([1,2,1], 12)
function([1,2,1]*9, '1-2')
function([1,'2',1]*9, 12)
function([1,2,0]*9, 12  )
function([1,2,1]*9, 12, p=None)
function([1,2,0]*9, 12, p=10)
```

````python
def function(l: list, s, **args):
    last_element = l[-1]        # last element (used as divisor)

    l[int(s)] = 10              # may raise IndexError or ValueError
    any_element = l[int(s) + 10]
    l[s] = 10                   # may raise TypeError if s is not int

    res = sum(l)                # may raise TypeError if list has non-numbers

    p = args['p']               # may raise KeyError if p not passed
    return res / last_element * p + any_element   # may raise ZeroDivisionError


test_cases = [
    lambda: function([1, 2, 1], 12),
    lambda: function([1, 2, 1] * 9, '1-2'),
    
    lambda: function([1, '2', 1] * 9, 12),
    lambda: function([1, 2, 0] * 9, 12),
    lambda: function([1, 2, 1] * 9, 12, p=None),
    lambda: function([1, 2, 0] * 9, 12, p=10)
]

for i, test_call in enumerate(test_cases, start=1):
    print(f"--- Test Case {i} ---")
    try:
        result = test_call()
        print(f"Success! Result: {result}")
    except Exception as e:
        print(f"Caught Error [{type(e).__name__}]: {e}")   # print error type + message
    print()
````

List have elemnts as any no of  `key-pair dict with key as list index and value as any integer`, `integers` and `numeric-strings`. There is always only one element in the dict.


```
l = [{0:2},2,3,4,'5', {5:10}]
```

````python

l = [{0:2}, 2, 3, 4, '5', {5:10}]
# For calculating sum of above list
s = 0
for i in range(len(l)):
    # You can Edit code from here
    try:
        item = l[i]
        
        if isinstance(item, dict):
            # Since there is always only one element in the dict, 
            # we can safely extract its only value
            s += list(item.values())[0]
            
        elif isinstance(item, (int, float)):
            s += item
            
        elif isinstance(item, str):
            s += int(item)
            
    except (ValueError, TypeError, IndexError) as e:
        # Handles unexpected data safely.
        print(f"Skipping index {i} due to an error: {e}")

print(s)
````

### `Q-3:`: File Handling with Exception handling

Write a program that opens a text file and write data to it as "Hello, Good Morning!!!". Handle exceptions that can be generated during the I/O operations. Do not show the success message on the main exception handling block (write inside the else block).

````python

try:
    file = open("morning.txt", "w")
    file.write("Hello, Good Morning!!!")

except FileNotFoundError as e:
    print(f"FileNotFoundError: {e}")    # path or directory does not exist

except PermissionError as e:
    print(f"PermissionError: {e}")      # no write access to the file

except IsADirectoryError as e:
    print(f"IsADirectoryError: {e}")    # given name is a directory not a file

except OSError as e:
    print(f"OSError: {e}")              # disk full or other OS level error

else:
    print("File written successfully!")
    file.close()                        # close only if write was successful
````

### `Q-4`: Number game program.

Write a number game program. Ask the user to enter a number. If the number is greater than number to be guessed, raise a **ValueTooLarge** exception. If the value is smaller the number to be guessed the, raise a **ValueTooSmall** exception and prompt the user to enter again. Quit the program only when the user enters the correct number. Also raise **GuessError** if user guess a number less than 1.

````python
# Custom exceptions for the number guessing game

class ValueTooLarge(Exception):
    pass    # raised when guess is too high

class ValueTooSmall(Exception):
    pass    # raised when guess is too low

class GuessError(Exception):
    pass    # raised when guess is out of valid range
````

````python
import random

secret = random.randint(1, 100)  # generate secret number between 1 and 100

while True:
    try:
        num = int(input("Guess the number (1-100): "))

        if num < 1:
            raise GuessError(f"{num} is less than 1!")      # below valid range

        if num > secret:
            raise ValueTooLarge(f"{num} is too large!")     # guess is too high

        if num < secret:
            raise ValueTooSmall(f"{num} is too small!")     # guess is too low

    except ValueError:
        print("Please enter a valid integer.\n")

    except GuessError as e:
        print(f"GuessError: {e}\n")

    except ValueTooLarge as e:
        print(f"ValueTooLarge: {e}\n")

    except ValueTooSmall as e:
        print(f"ValueTooSmall: {e}\n")

    else:
        print(f"Correct! The number was {secret}.")
        break                                               # exit on correct guess
````

### `Q-5:` Cast vote

Write a program that validate name and age as entered by the user to determine whether the person can cast vote or not. To handle the age, create **InvalidAge** exception and for name, create **InvalidName** exception. The name will be invalid when the string will be empty or name has only one word.

Example 1:

Input:

```bash
Enter the name:               Sohail Arif
Enter the age: 20
```

Output:

```bash
Sohail Arif  Congratulation !!! You can vote.
```

````python
# Custom exceptions for vote validation

class InvalidAge(Exception):
    pass    # raised when age is underage or unrealistic

class InvalidName(Exception):
    pass    # raised when name is empty or single word
````

````python
try:
    name = input("Enter the name: ").strip()
    age  = int(input("Enter the age:  "))

    if not name:
        raise InvalidName("Name cannot be empty!")                              # empty string

    if len(name.split()) < 2:
        raise InvalidName(f"'{name}' is invalid — full name required!")         # single word name

    if age < 18:
        raise InvalidAge(f"Age {age} is too young — must be 18 or above!")     # underage

    if age > 120:
        raise InvalidAge(f"Age {age} is not a valid age!")                      # unrealistic age

except ValueError:
    print("ValueError: Age must be a valid integer.")   # non-numeric age input

except InvalidName as e:
    print(f"InvalidName: {e}")

except InvalidAge as e:
    print(f"InvalidAge: {e}")

else:
    print(f"\n{name} — Congratulation!!! You can vote.")    # runs only if no exception raised
````

### `Q-6`: Write a python function which infinitely prints natural numbers in a single line. Raise the **StopIteration** exception after displaying first 20 numnbers to exit from the program.

````python
def print_natural_numbers():
    n = 1
    while True:
        try:
            if n > 20:
                raise StopIteration("Stopped after 20 numbers!")  # stop after 20
            print(n, end=" ")   # print on single line
            n += 1
        except StopIteration as e:
            print(f"\nStopIteration: {e}")
            break               # exit the loop


print_natural_numbers()
````
