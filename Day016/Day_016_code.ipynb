### `Q-2:` Write a Pandas program to add, subtract, multiple and divide two Pandas Series.

````python
import pandas as pd

# Create two Pandas Series
series1 = pd.Series([2, 4, 6, 8, 10])
series2 = pd.Series([1, 2, 3, 4, 5])

# Addition
addition = series1 + series2
print("Addition:")
print(addition)

# Subtraction
subtraction = series1 - series2
print("\nSubtraction:")
print(subtraction)

# Multiplication
multiplication = series1 * series2
print("\nMultiplication:")
print(multiplication)

# Division
division = series1 / series2
print("\nDivision:")
print(division)
````

### `Q-3` Write a Pandas program to compare the elements of the two Pandas Series.

````python
# Create two Pandas Series
series1 = pd.Series([2, 4, 6, 8, 10])
series2 = pd.Series([1, 3, 5, 7, 10])

# Equal to
print("Equal:")
print(series1 == series2)

# Greater than
print("\nGreater than:")
print(series1 > series2)

# Less than
print("\nLess than:")
print(series1 < series2)
````

### `Q-4.`Write a function to change the data type of given a column or a Series. Function takes series and data type as input, returns the converted series.
```
series = pd.Series([1,2,'Python', 2.0, True, 100])
change to float type data
```
Note: Read about pd.to_numeric()

````python
import pandas as pd

def change_dtype(series, dtype):
    """
    Function to change the data type of a given Series.
    
    Parameters:
    series (pd.Series): Input series
    dtype (str): Target data type
    
    Returns:
    pd.Series: Converted series
    """
    return pd.to_numeric(series, errors='coerce').astype(dtype)

# Sample Series
series = pd.Series([1, 2, 'Python', 2.0, True, 100])

print("Original Series:")
print(series)
print("Original dtype:", series.dtype)

# Convert to float
converted_series = change_dtype(series, 'float')

print("\nConverted Series:")
print(converted_series)
print("New dtype:", converted_series.dtype)
````

### `Q-8`
    i. Read `items.csv` making `item_name` as index.
    ii. Show no of nan values
    ii. Item price is given in $, so convert it to rupees without currency symbol.
    iii. Make data type of newly made series as float.
    iv. Fill nan with mean of the series


How csv file looks

```
item_name	item_price
Chips and Fresh Tomato Salsa	$2.39
Izze	$3.39
Nantucket Nectar	$3.39
Chips and Tomatillo-Green Chili Salsa	$2.39
Chicken Bowl	$16.98

```

````python
# i. Read items.csv making item_name as index, squeeze to Series
items_series = pd.read_csv("items.csv", index_col='item_name').squeeze()

# ii. Show number of NaN values
print(f"Total NaN values: {items_series.isna().sum()}")

# iii. Convert price from $ to rupees (remove currency symbol)
exchange_rate = 273  # 1 USD = 273 PKR (approx)
items_series = items_series.str.replace('$', '', regex=False).str.strip()
items_series = items_series.astype(float) * exchange_rate

# iv. Make data type as float (already done above, but confirming)
items_series = items_series.astype(float)
print(f"\nData type: {items_series.dtype}")

# v. Fill NaN with mean of the series

mean_price = items_series.mean()
items_series = items_series.fillna(mean_price)

print(f"\nMean Price (before fill): {mean_price:.2f}")
print(f"\nFinal Series:\n{items_series.head(10)}")
print(f"\nNaN values after fill: {items_series.isna().sum()}")
````

### `Q-9`:
    i. Find mean price
    ii. Find 30th and 6th percentile value
    iii. Plot Histogram on price with bin size 50
    iv. No of items price lies between [1000 to 2000]



````python
import matplotlib.pyplot as plt

# i. Mean price
mean_price = items_series.mean()
print(f"Mean Price: {mean_price:.2f}")

# ii. 30th and 60th percentile
percentile_30 = items_series.quantile(0.30)
percentile_60 = items_series.quantile(0.60)
print(f"30th Percentile: {percentile_30:.2f}")
print(f"60th Percentile: {percentile_60:.2f}")

# iii. Histogram with bin size 50
plt.figure(figsize=(10, 6))
plt.hist(items_series, bins=50, edgecolor='black')
plt.xlabel("Price (Rupees)")
plt.ylabel("Frequency")
plt.title("Distribution of Item Prices")
plt.savefig("histogram.png")
plt.show()

# iv. Items priced between 1000 and 2000
mask = (items_series >= 1000) & (items_series <= 2000)
count_in_range = mask.sum()
print(f"\nItems priced between 1000-2000: {count_in_range}")
````
