### `Q-6:` Make a dataframe of each team in IPL with details like - Team Name, Matches Played, Win%, Home Win%, Away Win%.
Show sorted dataframe on Win%

Replace old team name as new name before performing any tasks.
```
Delhi Daredevils ->Delhi Capitals
Kings XI Punjab -> Punjab Kings
Rising Pune Supergiants -> Rising Pune Supergiant
```

Note: Team1 represents Home team. Exclude not result matches.


````python
ipl_matches = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRy2DUdUbaKx_Co9F0FSnIlyS-8kp4aKv_I0-qzNeghiZHAI_hw94gKG22XTxNJHMFnFVKsO4xWOdIs/pub?gid=1655759976&single=true&output=csv")

print(ipl_matches.shape)
print(ipl_matches.columns.tolist())
print(ipl_matches.head(3))

````

````python
# Performing data analysis on the IPL dataset
# rename old team names
rename_map = {
    'Delhi Daredevils'       : 'Delhi Capitals',
    'Kings XI Punjab'        : 'Punjab Kings',
    'Rising Pune Supergiants': 'Rising Pune Supergiant',
}
for col in ['Team1', 'Team2', 'TossWinner', 'WinningTeam']:
    ipl_matches[col] = ipl_matches[col].replace(rename_map)

# exclude no result matches
df = ipl_matches[ipl_matches['WinningTeam'].notna()]

print(df.shape)
print(df['WinningTeam'].unique())
````

````python
# all unique teams
teams = pd.concat([df['Team1'], df['Team2']]).unique()
print("Total Teams:", len(teams))
print(teams)
````

````python
# step — calculate stats for each team
result = []

for team in teams:
    # all matches of this team
    matches = df[(df['Team1'] == team) | (df['Team2'] == team)]
    
    # home matches (Team1 = home)
    home = df[df['Team1'] == team]
    
    # away matches (Team2 = away)
    away = df[df['Team2'] == team]
    
    # wins
    total_wins = (matches['WinningTeam'] == team).sum()
    home_wins  = (home['WinningTeam'] == team).sum()
    away_wins  = (away['WinningTeam'] == team).sum()
    
    # percentages
    win_pct      = round(total_wins / len(matches) * 100, 2)
    home_win_pct = round(home_wins  / len(home)    * 100, 2) if len(home) > 0 else 0
    away_win_pct = round(away_wins  / len(away)    * 100, 2) if len(away) > 0 else 0
    
    result.append({
        'Team'         : team,
        'Matches Played': len(matches),
        'Win%'         : win_pct,
        'Home Win%'    : home_win_pct,
        'Away Win%'    : away_win_pct
    })

# convert to dataframe and sort by Win%
result_df = pd.DataFrame(result).sort_values('Win%', ascending=False)
result_df.reset_index(drop=True, inplace=True)
result_df.set_index('Team', inplace=True)
print(result_df)
````

````python
# filter final matches
finals = ipl_matches[ipl_matches['MatchNumber'] == 'Final']

print(finals.shape)
print(finals['Team1Players'].iloc[0])
print(type(finals['Team1Players'].iloc[0]))
````

````python
# combine Team1Players and Team2Players
all_players = pd.concat([finals['Team1Players'], finals['Team2Players']])

# clean and split — remove brackets and quotes, then split by comma
players_list = []
for row in all_players:
    # remove brackets and quotes
    row = row.strip("[]").replace("'", "")
    # split by comma and strip spaces
    players = [p.strip() for p in row.split(',')]
    players_list.extend(players)

# convert to series and count
players_series = pd.Series(players_list)
print(players_series.value_counts().head(10))
````

### `Q-9:` IPL Point Table

Make a function `point_table` which take `season` as parameter and show points table in non-ascendng order of points and in ascending order of team name.

For winning - 2 Ponits;

For loosing - 0 Point


For not result both team gets 1 points.

Make dataframe which will have
`TeamName` `MatchesPlayed` `MatchesWon` `NoResult` `Points`
make `TeamName` as index.

```
season parametr should be one of these->
['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
```


Output of two Top 2 in season 2022
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points

Gujarat Titans	    16	   12	       0	     24
Rajasthan Royals	17	   10	       0	     20

```

````python
print(ipl_matches['Season'].unique())
print(ipl_matches['Season'].dtype)
````

````python
def point_table(season):
    # filter by season
    season_df = ipl_matches[ipl_matches['Season'] == season]
    
    # all unique teams
    teams = pd.concat([season_df['Team1'], season_df['Team2']]).unique()
    
    result = []
    for team in teams:
        # all matches of this team
        matches = season_df[(season_df['Team1'] == team) | (season_df['Team2'] == team)]
        
        # no result matches
        no_result = matches[matches['WinningTeam'].isna()]
        
        # won matches
        won = (matches['WinningTeam'] == team).sum()
        
        # points
        points = (won * 2) + len(no_result)
        
        result.append({
            'TeamName'      : team,
            'MatchesPlayed' : len(matches),
            'MatchesWon'    : won,
            'NoResult'      : len(no_result),
            'Points'        : points
        })
    
    # convert to dataframe
    result_df = pd.DataFrame(result)
    
    # set index
    result_df.set_index('TeamName', inplace=True)
    
    # sort by Points descending, then TeamName ascending
    result_df = result_df.sort_values(['Points', 'TeamName'], 
                                       ascending=[False, True])
    return result_df

# call the function
print(point_table('2022'))
````

### `Q-10:` IPL Point Table cont.
Extend the above IPL Point Table with an extra column as `SeasonPosition`

Team below top 4 after sorting on `Points` and then on `TeamName` Will have same `SeasonPosition` as there rank. use rank function.


Teams in Top four will have `SeasonPosition` as:
```
    'Winner' - Team won final
    'Runner' - Team lost Final
    3 - Losing Team in Qualifier2
    4 - Losing Team in Eliminator
```

For changing value of pariticular cell use `df.at[row_index, col_label] = value`

Output of two Top 2 in season 2022. Your result should have all teams.
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points   SeasonPosition

Rajasthan Royals	  17	   10	       0	     20         Runner

```

Note: If you try to chnage value of view of any dataframe a warnig will be shown. To avoid it, make a copy of the dataframe you want to change in by `df.copy()`

````python
# check match types
print(ipl_matches['MatchNumber'].unique())
````

````python
# filter 2022 season special matches
season = '2022'
season_df = ipl_matches[ipl_matches['Season'] == season]

# final match
final = season_df[season_df['MatchNumber'] == 'Final']

# qualifier 2
qual2 = season_df[season_df['MatchNumber'] == 'Qualifier 2']

# eliminator
elim = season_df[season_df['MatchNumber'] == 'Eliminator']

print("Final:\n", final[['Team1', 'Team2', 'WinningTeam']])
print("\nQualifier 2:\n", qual2[['Team1', 'Team2', 'WinningTeam']])
print("\nEliminator:\n", elim[['Team1', 'Team2', 'WinningTeam']])
````

````python
def point_table(season):
    season_df = ipl_matches[ipl_matches['Season'] == season]
    
    # all unique teams
    teams = pd.concat([season_df['Team1'], season_df['Team2']]).unique()
    
    result = []
    for team in teams:
        matches   = season_df[(season_df['Team1'] == team) | (season_df['Team2'] == team)]
        no_result = matches[matches['WinningTeam'].isna()]
        won       = (matches['WinningTeam'] == team).sum()
        points    = (won * 2) + len(no_result)
        
        result.append({
            'TeamName'      : team,
            'MatchesPlayed' : len(matches),
            'MatchesWon'    : won,
            'NoResult'      : len(no_result),
            'Points'        : points
        })
    
    # Create a copy of the DataFrame to avoid warnings.
    result_df = pd.DataFrame(result).copy()
    result_df.set_index('TeamName', inplace=True)
    result_df = result_df.sort_values(['Points', 'TeamName'], ascending=[False, True])
    
    # Season position — using rank
    result_df['SeasonPosition'] = result_df['Points'].rank(method='first', ascending=False).astype(int)
    
    # special matches
    final = season_df[season_df['MatchNumber'] == 'Final'].iloc[0]
    qual2 = season_df[season_df['MatchNumber'] == 'Qualifier 2'].iloc[0]
    elim  = season_df[season_df['MatchNumber'] == 'Eliminator'].iloc[0]
    
    # # winner and runner
    winner = final['WinningTeam']
    runner = final['Team1'] if final['Team2'] == winner else final['Team2']
    
    # qualifier 2 loser → position 3
    qual2_loser = qual2['Team1'] if qual2['Team2'] == qual2['WinningTeam'] else qual2['Team2']
    
    # eliminator loser → position 4
    elim_loser = elim['Team1'] if elim['Team2'] == elim['WinningTeam'] else elim['Team2']
    
    # # set top 4 positions
    # convert SeasonPosition to object dtype first
    result_df['SeasonPosition'] = result_df['SeasonPosition'].astype(object)

    # Now you can assign string values.
    result_df.at[winner,      'SeasonPosition'] = 'Winner'
    result_df.at[runner,      'SeasonPosition'] = 'Runner'
    result_df.at[qual2_loser, 'SeasonPosition'] = 3
    result_df.at[elim_loser,  'SeasonPosition'] = 4
    
    return result_df
````
