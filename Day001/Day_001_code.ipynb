### Q2:- Write a program that will convert celsius value to fahrenheit.

````python
# Write your code here

celsius = int(input("Enter temperature in °Celcius: "))

fahrenheit = (celsius * 1.8) + 32    # Formula for converting Celsius to Fahrenheit.

print(f"{celsius}°C in Fahrenheit is: {fahrenheit}°F")     # Using f-string for formatted output. 
````

### Q3:- Take 2 numbers as input from the user.Write a program to swap the numbers without using any special python syntax.

````python
# Write your code here
a = int(input("Enter number 1: "))
b = int(input("Enter number 2: "))
print("<--Before swaping-->")
print(f"a -> {a} , b -> {b}")
temp = a           # Swap two numbers using a temporary variable.
a = b
b = temp

print("<--After swaping-->")       
print(f"a -> {a} , b -> {b}")  
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here
x1 = int(input("Enter x1: "))
y1 = int(input("Enter y1: "))

x2 = int(input("Enter x2: "))
y2 = int(input("Enter y2: "))

distance = ((x2 - x1)**2 + (y2 - y1)**2)**0.5   # Calculate Euclidean distance using the distance formula.

print(f"Distance: {distance}")
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here

heads = int(input("Enter heads: "))
legs = int(input("Enter legs: "))

dogs = (legs - heads*2)/2   # Calculate the number of dogs and chickens based on total heads and legs.

chickens = heads - dogs    

print(f"Dogs:{dogs}")
print(f"Chickens: {chickens}")
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
n = int(input("Enter a number: "))

square_sum = 0
for i in range(1,n+1):
    square_sum += i**2      # Calculate the sum of squares of the first n natural numbers.

print(f"Square Sum: {square_sum}")
    
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
a1 = int(input("Enter the first term (a1): "))
a2 = int(input("Enter the second term (a2): "))
n = int(input("Enter the term number (n) to find: "))

d = a2 - a1
        
# Calculate nth term
nth_term = a1 + (n - 1) * d   # Using the arithmetic sequence formula: a_n = a1 + (n-1)d
        
print(f"\nThe {n}th term of the series ({a1},{a2},{a2+d}... )is: {nth_term}")
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Write your code here
import math

# Tank dimensions (cuboid)
H = float(input("Enter tank Height (cm): "))
L = float(input("Enter tank Length (cm): "))
B = float(input("Enter tank Breadth (cm): "))

# Glass dimensions (cylinder)
h = float(input("Enter glass Height (cm): "))
r = float(input("Enter glass Radius (cm): "))

# Volume calculations
tank_volume  = H * L * B          # V = H × L × B
glass_volume = math.pi * r**2 * h # V = π r² h          instead of math.py we can also write '3.14159'


if glass_volume > 0:
    glasses = int(tank_volume // glass_volume)
    
    print(f"\nTank volume  : {tank_volume:.2f} cm³")  # Format output to 2 decimal places.
    print(f"Glass volume : {glass_volume:.2f} cm³")
    print(f"You can fill approximately {glasses} full glasses of milk.")
else:
    print("\nError: Glass dimensions must be greater than zero.")
````
