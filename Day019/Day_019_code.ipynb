### `Q-1:` Using `groupby` make groups using the `"Pclass"` column and find out the average age and total number of missing values in the `"Age"` column for every group.

````python
import pandas as pd

url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQjh5HzZ1N0SU7ME9ZQRzeVTaXaGsV97rU8R7eAcg53k27GTstJp9cRUOfr55go1GRRvTz1NwvyOnuh/pub?gid=1562145139&single=true&output=csv"
titanic_df = pd.read_csv(url)
titanic_df.head()
````

````python
# Group data by passenger class
grouped = titanic_df.groupby("Pclass")["Age"]

# Calculate the average age for each class
avg_age = grouped.mean()

# Count missing age values in each class
missing_age = grouped.apply(lambda x: x.isnull().sum())

# Create a summary DataFrame
result = pd.DataFrame({
    "avg_age": avg_age,
    "missing_age_count": missing_age
})

print(result)
````

### `Q-2:` Using `groupby` make groups using the `"Pclass"` column and fill every group's `"Embarked"` column's missing values with the mode value of that group. After that, print every group's `"Embarked"` column's value counts in ascending order.

````python
titanic_df.head()
````

````python
# fill missing "Embarked" with mode of each Pclass group
titanic_df["Embarked"] = titanic_df.groupby("Pclass")["Embarked"].transform(
    lambda x: x.fillna(x.mode()[0])
)

# print value counts (ascending) for each group
for pclass, group in titanic_df.groupby("Pclass"):
    print(f"Pclass {pclass}:")
    print(group["Embarked"].value_counts().sort_values(ascending=True))
    print()
````

### `Q-3:` Make groups based on `"Embarked"` column. And for each of this embarked group, make another group based on `"Pclass"` and find out the average fare (round off up to 2 decimal places) for each "Pclass" for each group of "Embarked".

**Sample Output:**

```bash
{'C': {1: 105, 2: 25, 3: 11},
 'Q': {1: 90, 2: 12, 3: 11},
 'S': {1: 70, 2: 20, 3: 15}}
```

````python
# Group data by embark location and passenger class
grouped = titanic_df.groupby(["Embarked", "Pclass"])["Fare"]

# Calculate the average fare
avg_fare = grouped.mean().round(2)

# Convert the result into table format
result_df = avg_fare.unstack()

print(result_df)
````

### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
import pandas as pd

fifa_df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vT3D_x_4DS6d51LKJ7ze1sxT5WpV5uiSVOFYHLwBiGru6vFyVv5h5-83AwFjxWYiWfCDjDAaarHAV-k/pub?gid=0&single=true&output=csv")
pd.set_option("display.max_columns", None)
fifa_df.head(2)
````

````python
def z_normalization(group, cols_to_perform):
    for col in cols_to_perform:
        mean = group[col].mean()      # mean of the group
        std = group[col].std()        # std deviation of the group
        group[col + "_zscore"] = (group[col] - mean) / std   # new zscore column
    return group
````

````python
grouped = fifa_df.groupby("Team")
# columns to normalize
cols_to_perform = ["Passes", "Passes Completed", "Attempted Line Breaks", "Completed Line Breaks"]

# group by team and apply the function
normalized_df = grouped.apply(lambda x: z_normalization(x, cols_to_perform))

# check new columns got added
normalized_df.head()
````

````python
# Calculate summary statistics for each team
result = grouped.agg({
    "Passes": ["min", "max"],
    "Yellow Cards": ["min", "max", "mean"],
    "Attempted Line Breaks": ["min", "max", "std"],
    "Possession (%)": "mean"
})

# Show all columns without wrapping
pd.set_option("display.expand_frame_repr", False)

print(result)
````

### `Q-5:` Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

````python
df = pd.read_csv("ipl_deliveries.csv")
df.head()
````

````python
df.head(2)
````

````python
# Filter only chasing innings
chasing_df = df[df["innings"] == 2]

# Calculate total runs and balls faced for each batter
batsman_stats = chasing_df.groupby("batter").agg(
    total_runs=("batsman_run", "sum"),
    balls_faced=("ballnumber", "count")
)

# Find the highest run scorer
highest_score = batsman_stats["total_runs"].idxmax()

# Calculate strike rate
batsman_stats["strike_rate"] = (
    batsman_stats["total_runs"] / batsman_stats["balls_faced"]
) * 100

# Keep batters who have faced at least 100 balls
eligible = batsman_stats[batsman_stats["balls_faced"] >= 100]

# Find the batter with the highest strike rate
best_strike_rate = eligible["strike_rate"].idxmax()

print("Highest scorer while chasing:", highest_score)
print("Best strike rate (100+ balls) while chasing:", best_strike_rate)
````

### `Q-7`: Most successful batting pair in IPL. Batting pair who have scored most runs playing together.


````python
df.head(2)
````

````python
# make a batting pair (order-independent)
df["pair"] = df.apply(lambda x: tuple(sorted([x["batter"], x["non-striker"]])), axis=1)

# total runs scored by each pair together
pair_runs = df.groupby("pair")["batsman_run"].sum().sort_values(ascending=False)

# pretty print top pair
top_pair, top_runs = pair_runs.index[0], pair_runs.iloc[0]
print(f"Successful Batting Pair: {top_pair[0]} & {top_pair[1]}")
print(f"Runs Scored Together: {top_runs}")
````

### `Q-8:` Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate.

````python
# pair (order-independent)
df["pair"] = df.apply(lambda x: tuple(sorted([x["batter"], x["non-striker"]])), axis=1)

# runs, balls faced (wides included as per note), dismissals for each pair
pair_stats = df.groupby("pair").agg(
    Runs=("batsman_run", "sum"),
    Balls=("ballnumber", "count"),
    Dismissals=("isWicketDelivery", "sum")
).reset_index()


# split pair tuple into two columns
pair_stats[["Batsman1", "Batsman2"]] = pd.DataFrame(pair_stats["pair"].tolist(), index=pair_stats.index)

# # average = runs per dismissal, strike rate = runs per 100 balls
pair_stats["Avg"] = (pair_stats["Runs"] / pair_stats["Dismissals"]).round(2)
pair_stats["StrikeRate"] = (pair_stats["Runs"] / pair_stats["Balls"] * 100).round(2)

# # final clean dataframe
final_df = pair_stats[["Batsman1", "Batsman2", "Runs", "Avg", "StrikeRate"]]

final_df.head()
````
