### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Write code here
annual_ctc = float(input("Enter your annual salary in lakhs: ")) * 100000

monthly_ctc = annual_ctc/12

deduction = monthly_ctc * (0.10 + 0.05 + 0.03)

if annual_ctc < 500000:
    annual_tax_rate = 0
elif annual_ctc <= 1000000:
    annual_tax_rate = 0.10
elif annual_ctc <= 2000000:
    annual_tax_rate = 0.20
else:
    annual_tax_rate = 0.30
    
monthly_tax = (annual_ctc * annual_tax_rate) / 12

# In-hand Salary
in_hand = monthly_ctc - deduction - monthly_tax

print(f"Your Monthly In-hand Salary is: {in_hand:.2f}")
````

### `Problem 3`: Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit.

````python
# Write code here
# Taking input from the user for Cost Price and Selling Price
cost_price = float(input("Enter the cost price: "))
selling_price = float(input("Enter the selling price: "))

# Comparing the prices to determine profit or loss
if selling_price > cost_price:
    profit = selling_price - cost_price
    print(f"It is a profit of {profit:.2f}")

elif cost_price > selling_price:
    loss = cost_price - selling_price
    print(f"It is a loss of {loss:.2f}")

else:
    print("There is no profit and no loss.")
````

### `Problem 4`: Write a menu-driven program -

````python
# Write code here
while True:
    # Display the menu options
    print("\n--- Unit Converter Menu ---")
    print("1. Centimeters to Feet")
    print("2. Kilometers to Miles")
    print("3. USD to PKR")
    print("4. Exit")
    
    choice = input("Enter your choice (1/2/3/4): ")
    
    # Handle the user's choice
    if choice == '1':
        cm = float(input("Enter length in cm: "))
        ft = cm * 0.0328084
        print(f"{cm} cm is equal to {ft:.2f} feet.")
        
    elif choice == '2':
        km = float(input("Enter distance in km: "))
        miles = km * 0.621371
        print(f"{km} km is equal to {miles:.2f} miles.")
        
    elif choice == '3':
        usd = float(input("Enter amount in USD: "))
        # Assuming 1 USD = 278 PKR as an example exchange rate
        pkr = usd * 278.0
        print(f"${usd} is equal to {pkr:.2f} PKR.")
        
    elif choice == '4':
        print("Exiting the program. Goodbye!")
        break # Exit the loop
        
    else:
        print("Invalid choice, please try again.")
````

### `Problem 5` - Exercise 12: Display Fibonacci series up to 10 terms.

*Note: The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13, 21. The next number in this series above is 13+21 = 34*

````python
# Write code here

# Number of terms to be printed in the Fibonacci series
n = 10

# Initialize the first two terms of the series: 0 and 1
a, b = 0, 1 
print(a, b, end=" ")

# Start the loop from 2 up to n, as the first two terms are already printed
for i in range(2, n):
    # Calculate the next term by adding the previous two
    temp = a + b
    print(temp, end=" ")
    
    # Update 'a' and 'b' for the next iteration:
    # 'a' takes the value of 'b'
    # 'b' takes the value of the newly calculated term (temp)
    a = b
    b = temp
````


````python
# Write code here

# First Method
# Taking integer input from the user
number = int(input("Enter an integer: "))

# Initialize the reversed number
reversed_number = 0

# Use a loop to reverse the digits
temp = number
while temp > 0:
    # Get the last digit
    digit = temp % 10
    
    # Build the reversed number
    reversed_number = (reversed_number * 10) + digit
    
    # Remove the last digit from the number
    temp = temp // 10

print(f"The reversed number is: {reversed_number}")

# Second Method
# convert int --> str then use s1[::-1] then comback to interger

str1 = str(number)
reversed = int(str1[::-1])

print(reversed)
````

### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
# Write code here
n = int(input("Enter a number: "))

total_sum = 0
current_number = 1

while current_number <= n:
    # Check if number is not divisible by 5
    if current_number % 5 != 0:
        # Check if adding the next number will exceed 300
        if total_sum + current_number > 300:
            break # Stop before adding the number
        total_sum += current_number
        
    current_number += 1

print(f"The final result is: {total_sum}")
````

### `Problem 9`: Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.

````python
# Write code here

# Initialize variables
total_sum = 0
count = 0

print("Enter numbers to calculate sum and average (Enter 0 to stop):")

while True:
    # Accept input from the user
    number = float(input("Enter a number: "))
    
    # Check if the input is zero to break the loop
    if number == 0:
        break
    
    # Add number to sum and increment the count
    total_sum += number
    count += 1

# Calculate and display the results
if count > 0:
    average = total_sum / count
    print(f"\nTotal Sum: {total_sum}")
    print(f"Average: {average:.2f}")
else:
    
    print("No numbers were entered.")
````

### `Problem 11`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
# Write code here


x ,y = 0,0
print("Enter commond (e.g UP 5) or '!' for exit")
while True:
    command = input()
    
    if command == "!":
        break
    
    parts = command.split()
    direction = parts[0]
    steps = int(parts[1])
    
    if direction.upper() == "UP":
        y += steps
    
    elif direction.upper() == "DOWN":
        y -= steps
        
    elif direction.upper() == "LEFT":
        x -= steps
    
    elif direction.upper() == "RIGHT":
        x += steps
    else:
        print("Invalid input.")
    
distance = (x**2 + y**2)**0.5

print(round(distance))
````

### `Problem 12`:Write a program to print whether a given number is a prime number or not

````python
# Write code here
n = int(input("Enter a number: "))

# Numbers less than 2 are not prime
if n < 2:
    print("Not a prime number.")
else:
    is_prime = True
    # Check for divisibility from 2 up to n-1
    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break # Once we find a divisor, we can stop the loop
    
    if is_prime:
        print("Yes, it is a prime number.")
    else:
        print("Not a prime number.")   
````

### `Problem 13`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
# Write code here

# Take the starting and ending range as input from the user[cite: 1]
start = int(input("Enter the starting number of the range: "))
end = int(input("Enter the ending number of the range: "))

print(f"\nArmstrong numbers between {start} and {end} are:")

# Loop through each number in the given range
for num in range(start, end + 1):
    # Convert number to string to find the total number of digits (power)
    num_str = str(num)
    power = len(num_str)
    
    sum_of_powers = 0
    # Calculate the sum of each digit raised to the power of the total number of digits
    for digit in num_str:
        sum_of_powers += int(digit) ** power
        
    # If the sum equals the original number, it is an Armstrong number[cite: 1]
    if sum_of_powers == num:
        print(num, end=" ")
    
````
