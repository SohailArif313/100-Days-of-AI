### **`Problem 3:`** Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.

```
Sample String : 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'
Expected Output :
No. of Upper case characters :  9
No. of Lower case Characters :  47
```

````python
# Write code here
s2 = 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'
lower_case = 0
upper_case = 0
for i in s2:
    if i.isupper():
        upper_case += 1
    if i.islower():
        lower_case += 1
        
print(f"No. of Upper case characters : {upper_case}\nNo. of Lower case characters : {lower_case}")
````

### **`Problem 5:`** Write a Python function to check whether a number is perfect or not.

A Perfect number is a number that is half the sum of all of its positive divisors (including itself).

Example :

```
The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1 + 2 + 3 = 6.
Equivalently, the number 6 is equal to half the sum of all its positive divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6.

The next perfect number is 28 = 1 + 2 + 4 + 7 + 14. This is followed by the perfect numbers 496 and 8128.
```


````python
# Write code here
# Find all perfect numbers between 1 and 10000
perfect_numbers = []

for i in range(1, 10000):

    # Find sum of all divisors except the number itself
    divisor_sum = sum(j for j in range(1, i) if i % j == 0)

    # If sum of divisors equals the number, it's a perfect number
    if divisor_sum == i:
        perfect_numbers.append(i)

print(perfect_numbers)
````

### **`Problem-7:`** Write a python function that accepts a string as input and returns the word with most occurence.

```
Input:
hello how are you i am fine thank you
```

```
Output
you -> 2
```

````python
# Write code here
# Input string
s = "hello how are you i am fine thank you"

# Split string into list of words
words = s.split()

# Get unique words using set
unique_words = set(words)

# Use map() to count occurrences of each unique word
word_count = dict(zip(unique_words, map(lambda word: words.count(word), unique_words)))

# Find word with maximum count
max_word = max(word_count, key=lambda x: word_count[x])

print(f"{max_word} -> {word_count[max_word]}")
````

### **`Problem-8:`** Write a python function that receives a list of integers and prints out a histogram of bin size 10

```
Input:
[13,42,15,37,22,39,41,50]
```

```
Output:
{11-20:2,21-30:1,31-40:2,41-50:3}
```

````python
((50-1) // 10) * 10 + 1
````

````python

lst = [13, 42, 15, 37, 22, 39, 41, 50]
bins = {}

for num in lst:
    
   # Calculate bin range (e.g. 13 -> 11-20)
    lower = ((num - 1) // 10) * 10 + 1
    upper = lower + 9

    # Create bin label
    bin_label = f"{lower}-{upper}"

    # If bin exists, increment count else set to 1
    if bin_label in bins:
        bins[bin_label] += 1
    else:
        bins[bin_label] = 1

    # Sort bins by lower bound value
        

sorted_bin = dict(sorted(bins.items(),key = lambda x:x[0].split("-")[0]))
print(sorted_bin)
````

### **`Problem-9:`** Write a python function that accepts a list of 2D co-ordinates and a query point, and then finds the the co-ordinate which is closest in terms of distance from the query point.

```
List of Coordinates

Query Point
(0,0)
```

```
Output
Nearest to (0,0) is (1,1)
```

````python
# Write code here
coordinates_list = [(1,1),(2,2),(3,3),(4,4)]
query_point = (0,0)

nearest_point = None
min_distance = float('inf')

for point in coordinates_list:
    distance = ((point[0]-query_point[0])**2 + (point[1] - query_point[1]) ** 2)**0.5
    if distance < min_distance:
        min_distance = distance
        coordinate_point = point
    
print(f"Nearest to {query_point} is {coordinate_point}")
    
````

### **`Problem-10`**:Write a python program that receives a list of strings and performs bag of word operation on those strings

https://en.wikipedia.org/wiki/Bag-of-words_model

````python
# Write code here
# Step 1: Build vocabulary of all unique words (sorted)
# Input sentences
sentences = ["I love cricket", "I love football", "I hate cricket"]

# Build sorted vocabulary of all unique words
vocabulary = sorted(set(word for sentence in sentences for word in sentence.split()))

# Print vocabulary
print("Vocabulary:", vocabulary)
print()

# Dictionary to store each sentence with its vector
whole_vector = {}

# Loop through each sentence
for sentence in sentences:

    # Split sentence into words
    words = sentence.split()

    # Create vector - 1 if word present, 0 if absent
    vector = [1 if word in words else 0 for word in vocabulary]

    # Store sentence and vector in dictionary
    whole_vector[sentence] = vector

# Print final result
print("Bag of Words:")
for sentence, vector in whole_vector.items():
    print(f"  {sentence} -> {vector}")
````
